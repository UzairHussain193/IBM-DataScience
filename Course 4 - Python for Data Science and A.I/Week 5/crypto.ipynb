```python
import pandas as pd
```

```python
start_time = 1686115200
end_time = 1759337599
```

```python
from pycoingecko import CoinGeckoAPI
cg=CoinGeckoAPI()
bitcoin_data = cg.get_coin_market_chart_range_by_id(id='bitcoin',
                                              vs_currency='usd',
                                              from_timestamp=start_time,
                                              to_timestamp=end_time
                                              )
# how to timestamp along with price and other data in coingecko api using python?
```
