```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('ChicagoCensusData.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
# 1. Standard Line Plot
plt.figure(figsize=(10, 6))
plt.plot(df['COMMUNITY_AREA_NUMBER'], df['HARDSHIP_INDEX'], label='Standard Line Plot')
plt.title('Standard Line Plot')
plt.xlabel('Community Area Number')
plt.ylabel('Hardship Index')
plt.legend()
plt.show()
```

```python
# 2. Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(df['PER_CAPITA_INCOME'], df['HARDSHIP_INDEX'], label='Scatter Plot', color='red')
plt.title('Scatter Plot')
plt.xlabel('Per Capita Income')
plt.ylabel('Hardship Index')
plt.legend()
plt.show()
```

```python
# 3. Histogram
plt.figure(figsize=(10, 6))
plt.hist(df['PERCENT_HOUSEHOLDS_BELOW_POVERTY'], bins=20, edgecolor='black', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Percent Households Below Poverty')
plt.ylabel('Frequency')
plt.show()
```

```python
# 4. Bar Plot
plt.figure(figsize=(10, 6))
plt.bar(df['COMMUNITY_AREA_NAME'], df['PERCENT_AGED_UNDER_18_OR_OVER_64'], color='green', alpha=0.7)
plt.title('Bar Plot')
plt.xlabel('Community Area Name')
plt.ylabel('Percent Aged Under 18 or Over 64')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# 5. Pseudo Color Plot
plt.pcolor(df.corr(), cmap='viridis')
plt.colorbar()
plt.title('Pseudo Color Plot (Correlation Matrix)')
plt.show()
```

```python
# Using Seaborn functions:

# 1. Regression Plot
sns.regplot(x='PER_CAPITA_INCOME', y='HARDSHIP_INDEX', data=df)
plt.title('Regression Plot')
plt.show()
```

```python
# 2. Box and Whisker Plot
sns.boxplot(x='PERCENT_AGED_UNDER_18_OR_OVER_64', y='PER_CAPITA_INCOME', data=df)
plt.title('Box and Whisker Plot')
plt.show()
```

```python
# 3. Residual Plot
sns.residplot(x='PERCENT_AGED_16__UNEMPLOYED', y='HARDSHIP_INDEX', data=df)
plt.title('Residual Plot')
plt.show()
```

```python
# 4. KDE Plot
sns.kdeplot(df['PER_CAPITA_INCOME'], shade=True)
plt.title('KDE Plot')
plt.show()
```

```python
# 5. Distribution Plot
sns.distplot(df['PERCENT_AGED_UNDER_18_OR_OVER_64'], hist=False, kde_kws={'shade': True})
plt.title('Distribution Plot')
plt.show()
```

```python
# 6. Regression Plot with Seaborn (continuation)
sns.regplot(x='PERCENT_AGED_UNDER_18_OR_OVER_64', y='PERCENT_HOUSEHOLDS_BELOW_POVERTY', data=df)
plt.title('Regression Plot with Seaborn')
plt.show()
```

```python
# 7. Box and Whisker Plot with Seaborn
sns.boxplot(x='PERCENT_AGED_25__WITHOUT_HIGH_SCHOOL_DIPLOMA', y='HARDSHIP_INDEX', data=df)
plt.title('Box and Whisker Plot with Seaborn')
plt.show()
```

```python
# 8. Residual Plot with Seaborn
sns.residplot(x='PERCENT_AGED_16__UNEMPLOYED', y='PERCENT_HOUSEHOLDS_BELOW_POVERTY', data=df)
plt.title('Residual Plot with Seaborn')
plt.show()
```

```python
# 9. KDE Plot with Seaborn
sns.kdeplot(df['PERCENT_AGED_25__WITHOUT_HIGH_SCHOOL_DIPLOMA'], shade=True)
plt.title('KDE Plot with Seaborn')
plt.show()
```

```python
# 10. Distribution Plot with Seaborn
sns.distplot(df['PER_CAPITA_INCOME'], hist=False, kde_kws={'shade': True})
plt.title('Distribution Plot with Seaborn')
plt.show()
```
