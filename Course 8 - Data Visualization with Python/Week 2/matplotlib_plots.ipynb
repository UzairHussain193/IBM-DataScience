# Creating dilrectly with matplotlib library

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create figure and axes
fig, ax = plt.subplots()
```

```python
# synthetic data to generate plots
years = np.arange(1980,2014)
immigrants = np.random.randint(2000,10000,size=(34,))
```

```python
# Line plot
fig, ax = plt.subplots()
ax.plot(years, immigrants)
# title the plot
plt.title("Immigrants between 1980 to 2013")
# labeling x axis
plt.xlabel("Years")
# labeling y axis
plt.ylabel("Total Immigrants")
# limits on x axis
plt.xlim(1975,2015)
# enabling grid
plt.grid(True)
# Legend
plt.legend(["immigrants"])
# customizing the appearence of the plot
ax.plot(years, immigrants,
        marker = 's',
        markersize = 5,
        color = 'green',
        linestyle = 'dotted')
# display plot
plt.show()
```

```python
# Scatter plot
fig, ax = plt.subplots()
ax.scatter(years,immigrants)
# title the plot
plt.title("Immigrants between 1980 to 2013")
# labeling x axis
plt.xlabel("Years")
# labeling y axis
plt.ylabel("Total Immigrants")
# limits on x axis
plt.xlim(1975,2015)
# enabling grid
plt.grid(True)
# Legend
plt.legend(["immigrants"])
# customizing the appearence of the plot
ax.scatter(years, immigrants,
        marker = '*',
        s = 20,
        color = 'red')
# legend at upper centre of the figure
plt.legend(["Immigrants"],loc='upper center')
# display plot
plt.show()
```

```python
# Bar plot
fig, ax = plt.subplots()
ax.bar(years,immigrants)
# title the plot
plt.title("Immigrants between 1980 to 2013")
# labeling x axis
plt.xlabel("Years")
# labeling y axis
plt.ylabel("Total Immigrants")
# display plot
plt.show()
```

```python
# Hist plot
fig, ax = plt.subplots()
ax.hist(immigrants, bins = 20, edgecolor='black',color='steelblue')
# title the plot
plt.title("Immigrants between 1980 to 2013")
# labeling x axis
plt.xlabel("Years")
# labeling y axis
plt.ylabel("Total Immigrants")
# display plot
plt.show()
```

```python
# Pie Chart
fig, ax = plt.subplots()
ax.pie(immigrants[0:5],labels=years[0:5],
       colors = ['gold','blue','lightgreen','coral','cyan'],
       autopct='%1.1f%%')
# title the plot
plt.title("Immigrants between 1980 to 2013")
# labeling x axis
plt.xlabel("Years")
# labeling y axis
plt.ylabel("Total Immigrants")
# display plot
plt.show()
```

```python
# multiple plots and subplotting
fig, axs = plt.subplots(1,2,sharey=True)
# plotting in the first aXes - the left one
axs[0].plot(years,immigrants)
axs[0].set_title("Line plot the immigrants")

# plotting in the second axes - the right one
axs[1].scatter(years,immigrants)
axs[1].set_title("Scatter plot the immigrants")

# display plot
plt.show()
```

```python
fig, axs = plt.subplots(1,2,sharey=True)

# adding the first subplot - top left
axs1 = fig.add_subplot(2,2,1)
# plotting in the first axes -  left one
axs1.plot(years,immigrants)
axs1.set_title("Plot 1 - the Line Plot")
# adding the second subplot - bottom left
axs2 = fig.add_subplot(2,2,2)
# plotting in the second axes -  right one
axs2.scatter(years,immigrants)
axs2.set_title("Plot 2 - the Scatter Plot")
# adding the third subplot - top right
axs3 = fig.add_subplot(2,2,3)
# plotting in the third axes -  left one
axs3.bar(years,immigrants)
axs3.set_title("Plot 3 - the Bar Plot")
# adding the fourth subplot - bottom right
axs4 = fig.add_subplot(2,2,4)
# plotting in the fourth axes -  right one
axs4.pie(immigrants[0:5],labels=years[0:5],
       colors = ['gold','blue','lightgreen','coral','cyan'],
       autopct='%1.1f%%')
axs4.set_aspect('equal')
axs4.set_title("Plot 4 - the Histogram Plot")

# display plot
plt.show()
```
